# SQL and Pandas

connect to data base through google drive.

```python
conn = sqlite3.connect('/content/drive/My Drive/ML_CityLit/world_development_indicators/database.sqlite')
```

## SQL Cursor

Cursor is used to execute statements to communicate with MySQL database.

```python
#creating a cursor we can use to navigate data 
cur = conn.cursor()

#in cur.execute we can put SQL code! 
# here we select everything that has the type data 
cur.execute("SELECT * FROM sqlite_master WHERE type='table'")

rows = cur.fetchall()

#we can then print the rows 
for row in rows:
    print(row)

#when you are done, you can close the cursos 
conn.close()
```

## or Pandas SQL

Perhaps the easiest way to use SQL here is to use it directly with pandas. We use pd.read_sql, and the sql command goes within three ivnerted commas. An SQL statement is composed of a sequence of keywords, identifiers, etc. terminated by a semicolon (;). The commands themselves are in CAPS

The most basic element of SQL is the Select element. The * symbol means all, so SELECT * is select all. In Pandas, we also specify our conenction to the database, which we here called conn.

```python
# This is so much easier to read than the previous cell 
# These are the names of all the individual tables inside the SQL database 


#The WHERE clause is used with the SELECT statement to extract only those records that fulfill specified conditions. The basic syntax can be given with:
#SELECT column_list FROM table_name WHERE condition;

pd.read_sql(""" SELECT *
                FROM sqlite_master
                WHERE type='table';""",
            conn)
```

### Select statements

```python
# just for fun, let us select everything from the table Indicators 
#Look at how many rows there are! 

pd.read_sql("""SELECT * FROM INDICATORS;""", conn)
```

```python
#We can also select from individual tables, here we select ten rows from the table Indicators 

pd.read_sql("""SELECT *
               FROM Indicators
               LIMIT 10;""",
           conn)
```

```python
# you can select columns in the data

pd.read_sql("""SELECT CountryCode, IndicatorName
               FROM Indicators;""",
           conn)
```

```python
#you can read everything from a table based on a column value

pd.read_sql("""SELECT *
               FROM Indicators
               WHERE CountryName = "Norway";""",
           conn)
```

### Filtering and ordering data

Can group data

```python
# here, we select the column region. We then count everything and label the result count. 

# We are selecting from the table 'country'
# then grouping by the column 'region', and ordering it by descending values in the 2nd column


pd.read_sql("""SELECT Region, COUNT(*) AS [Count]
               FROM Country 
               GROUP BY Region
               ORDER BY 2 DESC;""",
            conn)
```

AND &
Order

```python
# AND to specify several "wheres"
# we can also specify the ordering by the column name instead of the index (order by value)

pd.read_sql("""SELECT * 
               FROM Indicators
               WHERE IndicatorName='GDP per capita (current US$)'
               AND Year= 2010
               ORDER BY Value ASC
               LIMIT 10;""",
           conn)
```

The LIKE operator is used in a WHERE clause to search for a specified pattern in a column.

There are two wildcards often used in conjunction with the LIKE operator:
% represents zero, one, or multiple characters

_ underscore represents a single character


WHERE CustomerName LIKE 'a%' Finds any values that start with "a"

WHERE CustomerName LIKE '%a' Finds any values that end with "a"

WHERE CustomerName LIKE '%or%' Finds any values that have "or" in any position

WHERE CustomerName LIKE '_r%' Finds any values that have "r" in the second position

WHERE CustomerName LIKE 'a_%' Finds any values that start with "a" and are at least 2 characters in length

WHERE CustomerName LIKE 'a__%' Finds any values that start with "a" and are at least 3 characters in length

WHERE ContactName LIKE 'a%o' Finds any values that start with "a" and ends with "o"

```python
#no countries beginning with h

pd.read_sql("""SELECT *
               FROM Indicators
               WHERE CountryName NOT LIKE 'H%'
               ORDER BY YEAR DESC, IndicatorName 
               LIMIT 10;""",
           conn)
```

Union

```python
# The UNION operator is used to combine the results of two or more SELECT queries into a single result set
# If you're into maths, this is like a union in probability 

pd.read_sql("""SELECT *
               FROM Indicators
               WHERE (CountryName = "Chile" AND Year=2010)
               UNION SELECT *
               FROM Indicators
               WHERE (CountryName = "Peru" AND Year=2011);""",
           conn)
```

### Aggregating data

```python
# There are several built in functions to aggregate data, such as average

pd.read_sql("""SELECT AVG(Value)
               FROM Indicators
               WHERE IndicatorName = 'Urban population';""",
           conn)
```

### Join

```python
# A lot of times you will need to join data from different sources or tables
# There are many different ways to do this


pd.read_sql(""" SELECT Indicators.*, Series.LongDefinition
                FROM Indicators
                LEFT JOIN Series 
                ON Indicators.IndicatorName  = Series.IndicatorName
                WHERE Indicators.IndicatorName LIKE 'Fertility rate%'
                AND CountryName ='Bolivia';""",
            conn)
```

## Storing Variables
You can store your data frames in your google drive. It may be beneficial to you to do a lot of preprocessing, and then save the final result, so you don't have to re-run the preprocessing every single time. There are many ways to store these files. One of the Pandas ways is to store stuff as a pickle. This is a compressed format, so it will take up less space on your drive.

```python
result_df = pd.read_sql(""" SELECT Indicators.*, Series.LongDefinition
                FROM Indicators
                LEFT JOIN Series 
                ON Indicators.IndicatorName  = Series.IndicatorName
                WHERE Indicators.IndicatorName LIKE 'Fertility rate%'
                AND CountryName ='Bolivia';""",
            conn)

#name your pickle with extension .pkl
result_df.to_pickle("processedDataWithVerySpecificName.pkl")
```
